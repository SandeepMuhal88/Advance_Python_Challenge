```python
import matplotlib.pyplot as plt
import numpy as np
arr=np.arange(1,9)
print(arr)
arr2=(arr*2)
print(arr2)
```

```python
from matplotlib import pyplot as plt
import numpy as np
arr3=np.array([7,15,8,23,41,4,9,6])
plt.plot(arr,arr3)
plt.plot(arr,arr2,color='red',linestyle=':',linewidth=2)
plt.title("Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
arr = [1, 2, 3, 4, 5]
arr2 = [2, 4, 6, 8, 10]
arr3 = [1, 4, 9, 16, 25]
print(arr)
print(arr2)
print(arr3)
```

```python
plt.subplots(1,2,figsize=(10,5))
plt.plot(arr,arr3)
plt.subplot(1,2,1)
plt.plot(arr,arr2,color='red',linestyle=':',linewidth=2)
plt.title("Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=["Python","Java","C++","C#","JavaScript"]
y=[75,85,96,95,87]
plt.xlabel("Languages")
plt.ylabel("Percentage")
plt.title("Languages")
plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
arr=np.array(["Stdent 1","Stdent 2","Stdent 3","Stdent 4","Stdent 5"])
arr2=np.array([75,85,96,95,87])
plt.scatter(arr,arr2,color='red',marker='o')
plt.xlabel("Students")
plt.ylabel("Percentage")
plt.title("Student_details")
plt.show()
plt.xlabel("Students")
plt.ylabel("Percentage")
plt.title("Students")
plt.barh(arr,arr2)
plt.show()
```
