1. Write a program to swap two variables without using a third variable.

```python
def swap_number(a,b):
    print(f"The number : a={a},b={b}")
    a,b=b,a
    print(f"The swap number: a={a},b={b}")



swap_number(5,6)
    
```

2. Take a number as input and print whether it is positive, negative, or zero.

```python
def check_number(num):
    if num<0:
        print("Given Number is Negative !!")
    elif num>0:
        print("Given Number is Positive ###")
    else:
        print("Given Number is Zero ***")
    

check_number(2)
check_number(-7)
check_number(0)
```

3. Take two integers and print their sum, difference, product, and quotient.

```python
def operation(a, b):
    print("The sum of integer number :-", a + b)
    print("The difference of integer number :-", a - b)
    print("The product of integer number :-", a * b)
    # use float division for quotient (use '//' for integer division)
    print("The quotient of integer number :-", a / b)


operation(5, 7)
```
