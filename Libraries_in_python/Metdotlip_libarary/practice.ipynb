```python
import matplotlib.pyplot as plt

def plotFunction(a,b,step):
    nsteps = int((b-a)/step)
    x=[a+step*i for i in range(nsteps)]
    y1=[i**2 for i in x]
    y2=[i**3 for i in x]
    plt.plot(x,y1,label='x^2')
    plt.plot(x,y2,label='x^3')
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Plot of x^2 and x^3')
    plt.grid(True)
    plt.show()
```

```python
def main():
    a=float(input('Enter the start of the range: '))
    b=float(input('Enter the end of the range: '))
    step=float(input('Enter the step size: '))
    plotFunction(a,b,step)
    

if __name__ == '__main__':
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np
def plotHistogram():
    data = [5, 10, 15, 20, 25]
    plt.hist(data, bins=[0, 5, 10, 15, 20, 25])
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Histogram of data')
    plt.show()

def main():
    plotHistogram()

if __name__ == '__main__':
    main()
```

```python
#improve version
import matplotlib.pyplot as plt
import numpy as np

def AplotHistogram():
    data=[5, 10, 15, 20, 25, 25, 15, 10, 5, 5, 10, 20]
    num_bins=3
    bins=np.linspace(min(data),max(data),num_bins+1)
    plt.hist(data,bins=bins,color='skyblue',edgecolor='black',alpha=0.7)

    counts,edges=np.histogram(data,bins=bins)
    for i in range(num_bins):
        plt.text((edges[i]+edges[i+1])/2,counts[i],str(counts[i]),fontsize=12,horizontalalignment='center')
    plt.xlabel('Value Ranges (Bins)', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.title('Enhanced Histogram of Data', fontsize=14, fontweight='bold')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.xticks(bins, rotation=45)
    plt.tight_layout()
    plt.show()

def main():
    AplotHistogram()

if __name__ == '__main__':
    main()
```
