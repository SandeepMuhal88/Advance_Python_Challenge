```python
import time
import matplotlib.pyplot as plt
import random

# 1. Define the function you want to test
def find_element(l1, y):
    for i in l1:
        if i == y:
            return True
    return False

# 2. Prepare Data for Plotting
input_sizes = [1000, 5000, 10000, 50000, 100000, 200000] # Different 'N'
times = []

# 3. Run the Experiment
for n in input_sizes:
    # Generate a random list of size N
    data = list(range(n))
    target = -1 # Worst case: Searching for something that doesn't exist
    
    start_time = time.time()
    find_element(data, target)
    end_time = time.time()
    
    times.append(end_time - start_time)

# 4. Plot the Results
plt.figure(figsize=(10, 6))
plt.plot(input_sizes, times, marker='o', color='b', label='Actual Time')
plt.xlabel('Input Size (N)')
plt.ylabel('Time (Seconds)')
plt.title('Time Complexity Analysis')
plt.grid(True)
plt.show()
```
